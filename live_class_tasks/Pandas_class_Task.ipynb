```python
# 1 .Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave

week = pd.to_datetime(df['Date'])
df['date_format'] = week

df['week'] = df['date_format'].dt.isocalendar().week
df.groupby('week')['Average Rating'].mean()
```

```python
df.groupby(['Agent Name','week'])['Average Rating'].mean() 
```

```python
# 2 .Total working days for each agents 

df['date_day'] = df['date_format'].dt.day
df.groupby('Agent Name')['date_day'].count()
```

```python
# 9 . average weekly response time for each agent 

df['Average_response_time_seconds'] = df['Average Response Time'].str.split(':').apply(lambda x: int(x[1]) * 60 + int(x[2]))
df.groupby(['Agent Name', 'week'])['Average_response_time_seconds'].mean()
```

```python
# 10 . average weekely resolution time for each agents 

df['Average_resolution_time_minutes'] = df['Average Resolution Time'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]))
df.groupby(['Agent Name', 'week'])['Average_resolution_time_minutes'].mean()
```

```python
# 13 . Total contributation hour for each and every agents weekly basis 

week = pd.to_datetime(df1['Date'])
df1['date_format'] = week
df1['week'] = df['date_format'].dt.isocalendar().week

df1['Duration_Hours'] = df1['Duration'].str.split(':').apply(lambda x: int(x[0]))
df1.groupby(['Agent','week'])['Duration_Hours'].sum()
```
